```python
from lxml import etree as ET

data = ET.parse('combined.xml')
data
```

```python
persons=data.xpath('//dblpperson')
persons
```

```python
pids = [person.attrib['pid'] for person in persons]
```

```python
pid_dict = dict(zip(pids, names))
pid_dict
```

```python
index_dict = dict(zip(pids,num_index))
index_dict
inversed_index_dict = dict(zip(num_index,pids))
```

```python
#read all r labels
r_labels = data.xpath('//r')
r_labels
```

```python
#convert r labels into edge
edges= {}
cnt= 0
import tqdm
processbar = tqdm.tqdm(total=len(r_labels))
for r in r_labels:
    processbar.update(1)
    #find all authors
    authors = r.xpath('.//author')
    year = r.xpath('.//year')[0].text
    #if there is only one author, skip
    if len(authors) < 2:
        continue
    #get the author pids
    author_pids = [author.attrib['pid'] for author in authors]
    #create edges if pid is in pid_dict
    for i in range(len(author_pids)):
        for j in range(i+1, len(author_pids)):
            if author_pids[i] in pid_dict and author_pids[j] in pid_dict:
                cnt+=1
                if year not in edges:
                    edges[year] = []
                if (index_dict[author_pids[i]], index_dict[author_pids[j]]) not in edges[year] and (index_dict[author_pids[j]], index_dict[author_pids[i]]) not in edges[year]:
                    edges[year].append((index_dict[author_pids[i]], index_dict[author_pids[j]]))

for year in edges:
    edges[year] = list(set(edges[year]))
    print(year, len(edges[year]))
edges
    
```

```python
#save the index_dict
import json
with open('index_dict.json', 'w') as f:
    json.dump(index_dict, f)
#save the edges
with open('edges.json', 'w') as f:
    json.dump(edges, f)
#save the pid_dict
with open('pid_dict.json', 'w') as f:
    json.dump(pid_dict, f)
#save the inversed_index_dict
with open('inversed_index_dict.json', 'w') as f:
    json.dump(inversed_index_dict, f)
```

```python
#find its note label with type "affiliation"
# affiliations = persons[0].xpath('.//note[@type="affiliation"]')
# affiliations
special_case = ["Saudi Arabia", "Hong Kong", "New Zealand"]
def find_affiliation(person):
    p=person.find('.//person')
    affiliations = p.xpath('.//note[@type="affiliation"]')
    affs = []
    orgs= []
    countries = []
    if len(affiliations) == 0:
        return None
    else:
        for affiliation in affiliations:
            if affiliation.text is not None:
                affs.append(affiliation.text)
            #split the affiliation into org and country, by comma;
                try:
                    # org= affiliation.text.split(', ')[0]
                    # country = affiliation.text.split(', ')[-1]
                #if it without comma, it should be the org
                    if len(affiliation.text.split(', ')) == 1:
                        org = affiliation.text
                        country = None
                    else:
                        org= affiliation.text.split(', ')[0]
                        country = affiliation.text.split(', ')[-1]
                        if "The " in country:
                            country = country[4:]#remove "The "
                        if country not in special_case and len(country.split(' ')) > 1:
                            country = None
                except:
                    print(affiliation.text)
                    print(person.attrib['pid'])
                    print("NOOOOO")
                    org = None
                    country = None
            else:
                org = None
                country = None
            if org is not None:
                orgs.append(org)
                countries.append(country)
    affs = list(set(affs))
    orgs = list(set(orgs))
    countries = list(set(countries))
    #remove [None] 
    if None in orgs:
        orgs.remove(None)
    if None in countries:
        countries.remove(None)
    if len(orgs) == 0:
        orgs = None
    if len(countries) == 0:
        countries = None
    para={"orgs":orgs, "countries":countries}
    return  para
```

```python
country_dict = {}
org_dict = {}
for person in persons:
    para = find_affiliation(person)
    if para is None:
        country_dict[index_dict[person.attrib['pid']]] = None
        org_dict[index_dict[person.attrib['pid']]] = None
        continue
    orgs = para['orgs']
    countries = para['countries']
    pid= person.attrib['pid']
    idx = index_dict[pid]
    if orgs is not None:
        country_dict[idx] = countries
        org_dict[idx] = orgs
country_dict
```
